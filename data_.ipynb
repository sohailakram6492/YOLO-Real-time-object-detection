```python
import json , os
import cv2
```

```python
def get_images():
    name_img = os.listdir('./images')
    temp={}
    for i in name_img:
        temp[i]=cv2.imread(f'./images/{i}').copy()
    # print(temp.values())
    return temp
get_images()
```

```python
with open('via_project.json') as file:
  db = json.load(file)
filter = ['filename','x','y','width','height','cat','dog']
```

```python
def get_depth(dic):
    list_ = []
    
    def val(key, value):
        if key in filter or value in filter:
            list_.append(str(f'{key}:{value}'))

    def check_dic(dic):
        if isinstance(dic,dict):
            for k,v in dic.items():
                if isinstance(v, (dict, list)):
                    check_dic(v)
                else:
                    val(k,v)
        elif isinstance(dic , (list)):
            for d in dic:
                if isinstance(d, (dict, list)):
                    check_dic(d)
                # else:                 #  ?????
                #     val(k,v)

    check_dic(dic)
    return list_
main_data = get_depth(db)
main_data
```

```python
def normalize_for_yolo(xmin,ymin,w,h,w_img,h_img):
    xcenter = (xmin + w/2) / w_img
    ycenter = (ymin + h/2) / h_img
    w = w / w_img
    h = h / h_img
    return xcenter,ycenter,w,h
if not os.path.exists('./yolo_annotations'):
    os.mkdir('./yolo_annotations')
image=''
file=''
points=[]
images=get_images()
for i in main_data:
    # print(i)
    if i.split(':')[0] == 'filename':
        image=images[i.split(':')[1]]
        file=open(f'yolo_annotations/{i.split(":")[1].split(".")[0]}.txt','w')
    else:
        if i.split(':')[0] == 'name':
            # print(i)
            if i.split(':')[1]=='cat':
                cls=1
            if i.split(':')[1]=='dog':
                cls=0
            file.write(f"{cls}")
            x,y,w,h=[int(x) for x in points]
            # image=cv2.rectangle(image,(x,y),(x+w,y+h),[0,255,0],2)
            # image=cv2.putText(image,str(cls),(x,y),cv2.FONT_HERSHEY_SIMPLEX,2,[0,255,0],2)
            # cv2.imshow('',image)
            # cv2.waitKey(0)
            # cv2.destroyAllWindows()
            x,y,w,h=normalize_for_yolo(x,y,w,h,image.shape[1],image.shape[0])
            file.write(f"{x} {y} {w} {h}\n")
            points=[]
        else:
            points.append(i.split(":")[1])
file.close()
    
```
